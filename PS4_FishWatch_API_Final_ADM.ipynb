```python
#!/users/bin/env python

#Printing The Full List of Avaliable Fish to Choose from in the Database, scroll down to the next cell after getting the fish list

import urllib.request, urllib.parse, urllib.error
import json


serviceurl = 'https://www.fishwatch.gov/api/species'

while True:
    fishyfish = input('Enter anything to look at all fish in the database to find information for: ')
    if len(fishyfish) < 1: break

    url = serviceurl + "?" + urllib.parse.urlencode(
        {'fishyfish': fishyfish})

    print('Retrieving', url)
    uh = urllib.request.urlopen(urllib.request.Request(url,headers={'User-Agent': 'Mozilla/5.0'}))
    data = uh.read().decode()
    print('Retrieved', len(data), 'characters')

    try:
        js = json.loads(data)
    except:
        js = None

    for fishyfish in js:
        print("Fish list:", fishyfish['Species Name'])
    #print(data) # only needed if you need to print all the data
       
```
